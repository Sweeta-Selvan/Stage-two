```python
import decoupler as dc
```

```python
# fetch the PanglaoDB from omnipath
markers = dc.op.resource(name="PanglaoDB", organism= "human")
#keep only canonical cell type markers
markers = markers[markers["canonical_marker"]]
#remove duplicate cell type and genes
markers = markers[~markers.duplicated(["cell_type", "genesymbol"])]
#rename for decoupler compatibility
markers   = markers.rename(columns = {"cell_type" :"source", "genesymbol": "target"})
#keep only source and target
markers = markers[["source", "target"]]
markers.head()
```

```python
# Before running decoupler, ensure that the variable names (genes) in bm_adata match
# the gene symbols in the `markers` DataFrame.
# Currently, bm_adata.var_names are Ensembl IDs, and markers.target are gene symbols.
# We need to set bm_adata.var_names to the 'feature_name' column (gene symbols).
bm_adata.var_names = bm_adata.var['feature_name'].astype(str)

# Make variable names unique to handle potential duplicate gene symbols, a common requirement for scanpy/anndata operations.
bm_adata.var_names_make_unique()

# load our data to dc
dc.mt.ulm(data= bm_adata, net = markers, tmin= 3)
```

```python
# Example of how to subset for multiple genes in the 'source' column
available_genes = set(bm_adata.var_names)

Neutro_markers = markers[markers['source'].isin(['Neutrophils'])]['target']
Neutro_markers = Neutro_markers[Neutro_markers.isin(available_genes)]

B_cell_naive_markers = markers[markers['source'].isin(['B cells naive'])]['target']
B_cell_naive_markers = B_cell_naive_markers[B_cell_naive_markers.isin(available_genes)]

t_cells_markers = markers[markers['source'].isin(['T cells'])]['target']
t_cells_markers = t_cells_markers[t_cells_markers.isin(available_genes)]
```
